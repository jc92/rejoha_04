```python
from PIL import Image
from tqdm import  tqdm
import cv2
```

```python
# get frame
def get_frame(video,frame):
    video.set(cv2.CAP_PROP_POS_FRAMES, frame)
    ret, frame = video.read()
    imageRGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    im = Image.fromarray(imageRGB)
    return im

def generate_img(file_loc,start_frame):
    video = cv2.VideoCapture(file_loc)
    frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    w = int(video.get(cv2.CAP_PROP_FRAME_WIDTH ))
    h = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT ))
    print(f"video width {w}")
    print(f"video height {h}")
    print(f"video frames {frames}")

    output_image = Image.new("RGB", (frames,h))
    for i in tqdm(range(frames)):
        try:
            im = get_frame(video,frame=i)
            im_ = im.crop((start_frame, 0, start_frame+1, h))
            output_image.paste(im_,(i,0))
        except Exception as e:
            print(e)
    return output_image
```

```python
file_loc = './projects/test1/01_input/test1.mp4'
start_frame = 250
output = generate_img(file_loc,start_frame)
```

```python
import cv2
import numpy as np
from tqdm import tqdm
from PIL import Image

file_loc = './projects/fastnacht/01_input/fastnacht_short.mp4'
video = cv2.VideoCapture(file_loc)
w_video = int(video.get(cv2.CAP_PROP_FRAME_WIDTH ))
h_video = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT ))
slice_width = 3
frames_counter = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
# output_frame = Image.new("RGB", (frames_counter,h))
output_frame = []
with tqdm(total=frames_counter) as pbar:
    while True:
        read, frame= video.read()
        pbar.update(1)
        if not read:
            break
        h,w,c = frame.shape        # todo opt     
        new_line = np.array([x[0][::-1] for x in frame]).astype('uint8').reshape(h, -1, c)
        for _ in range(slice_width):
            output_frame.append(new_line)
#         for _ in range(slice_width):
#             read, frame= video.read()
#             pbar.update(1)
```
